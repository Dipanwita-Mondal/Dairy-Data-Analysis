### Importing the libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv(r"C:\Users\adaka\Downloads\dairy_dataset.csv")
```

```python
Features: 
Location:                           The geographical location of the dairy farm.
Total Land Area (acres):             The total land area occupied by the dairy farm.
Number of Cows:                      The number of cows present in the dairy farm.
Farm Size:                           The size of the dairy farm(in sq.km).
Date:                                The date of data recording.
Product ID:                          The unique identifier for each dairy product.
Product Name:                        The name of the dairy product.
Brand:                               The brand associated with the dairy product.
Quantity (liters/kg):                The quantity of the dairy product available.
Price per Unit:                      The price per unit of the dairy product.
Total Value:                         The total value of the available quantity of the dairy product.
Shelf Life (days):                   The shelf life of the dairy product in days.
Storage Condition:                   The recommended storage condition for the dairy product.
Production Date:                     The date of production for the dairy product.
Expiration Date:                     The date of expiration for the dairy product.
Quantity Sold (liters/kg):           The quantity of the dairy product sold.
Price per Unit (sold):               The price per unit at which the dairy product was sold.
Approx. Total Revenue (INR):         The approximate total revenue generated from the sale of the dairy product.
Customer Location:                   The location of the customer who purchased the dairy product.
Sales Channel:                       The channel through which the dairy product was sold (Retail, Wholesale, Online).
Quantity in Stock (liters/kg):       The quantity of the dairy product remaining in stock.
Minimum Stock Threshold (liters/kg): The minimum stock threshold for the dairy product.
Reorder Quantity (liters/kg):        The recommended quantity to reorder for the dairy product.
```

### Analysis

```python
#### Total Revenue of dairy products
revenue_total=df['Approx. Total Revenue(INR)'].sum()
print(f"Total Revenue(INR): {revenue_total}")
```

```python
product_names = df['Product Name'].unique()
print(product_names)
```

```python
quantity_sold_per_product=df.groupby(['Product Name'])['Quantity Sold (liters/kg)'].sum().sort_values(ascending=False)
quantity_sold_per_product
```

##### Maximum selling product

```python
# Plotting a pie chart
custom_palette = sns.color_palette("Set3", len(quantity_sold_per_product))
quantity_sold_per_product.plot(kind='pie', autopct='%1.1f%%', startangle=90,colors=custom_palette)
plt.title('Quantity Sold per Product')
plt.axis('equal')
plt.show()

max_selling_product = quantity_sold_per_product.idxmax()
print(f"Maximum selling product is : {max_selling_product}")
```

### Total Land area in acres by Brand

```python
brands=df.groupby(['Brand'])['Total Land Area (acres)'].sum().sort_values()
brands
```

```python
# Creating the horizontal bar chart
custom_palette = sns.color_palette("OrRd", len(brands))
plt.barh(brands.index, brands.values,color=custom_palette)

plt.xlabel('Total Land Area (acres)')
plt.ylabel('Brand')
plt.title('Total Land Area by Brand')
plt.show()
```

#### Ovservation: Amul has the maximum land area(acres) and Warana has the minimum land area.

#### Q. From which state maximum revenue of Amul is generated?

```python
amul = df[df['Brand'] == 'Amul']
state_top=amul.groupby(['Customer Location'])['Approx. Total Revenue(INR)'].sum().sort_values(ascending=False)
print(state_top)
x=state_top.index[0]
print(f"Amul has it's maximum revenue generated from:{x}")
```

#### Total revenue of each brand by location

```python
df.groupby(['Brand'])['Approx. Total Revenue(INR)'].sum().sort_values(ascending=False)
```

```python
total_revenue_by_brand = df.groupby('Brand')['Approx. Total Revenue(INR)'].sum()

plt.figure(figsize=(8, 8))
plt.pie(total_revenue_by_brand, labels=total_revenue_by_brand.index, autopct='%1.1f%%')
plt.title('Total Revenue by Brand')
plt.show()
```

#### Q. Which product has highest shelf life?

```python
shelf_life=df.groupby(['Product Name'])['Shelf Life (days)'].sum().sort_values(ascending=False)
shelf_life
```

```python
# Creating the horizontal bar chart
plt.barh(shelf_life.index, shelf_life.values,color='purple')

plt.xlabel('Shelf Life(days)')
plt.ylabel('Product')
plt.title('Product Vs Shelf life(days)')
plt.show()
```

#### Ovservation: Ghee has the maximum shelf life(days) and curd has the minimum shelf life(days).

#### Effect of Storage condition on Quantity in Stocks

```python
storage_of_product = df.groupby(['Storage Condition'])['Quantity in Stock (liters/kg)'].sum()
storage_of_product
```

```python
plt.figure(figsize=(8, 4))
storage_of_product.plot(kind='bar')
plt.xlabel('Storage Condition')
plt.ylabel('Quantity in Stock (liters/kg)')
plt.title('Quantity in Stock by Storage Condition')
plt.show()
```

#### Observation:
- Refrigerated products are kept in stock in larger quantity than other products.
- Products in Polythene Packet and Terapack are kept in lesser quantity in stock.

### Various sales channels in different parts of India

```python
location_df = df.groupby(['Location', 'Sales Channel'])['Sales Channel'].count()

# Pivot the data to create a DataFrame with 'Location' as columns
pivot_df = location_df.unstack()

# Plotting a stacked bar chart
pivot_df.plot(kind='bar', stacked=False)

plt.xlabel('Sales Channel')
plt.ylabel('Count')
plt.title('Sales Channel Count by Location')

plt.legend(title='Location')
plt.show()
```

#### Observation:
- In Delhi Retail sales channel is more, same with Haryana,Jharkhand,Telangana,West Bengal
- Online sales channel is more in Bihar,Maharashra and Rajasthan.
- Wholesale sales channel is more in Gujrat,Kerala and Madhya Pradesh.

  This information can be used to optimize marketing and distribution strategies.


### Stocks Deficiet summary

```python
df['Stock Deficit'] = df['Reorder Quantity (liters/kg)'] - df['Quantity in Stock (liters/kg)']
products_to_reorder = df[df['Stock Deficit'] > 0]
products_to_reorder_subset = products_to_reorder[['Product Name', 'Brand','Stock Deficit']]
products_to_reorder_subset 
```

```python
reorder_summary = products_to_reorder_subset.groupby(['Product Name', 'Brand'])['Stock Deficit'].sum()
print(reorder_summary)
```

### Correlation Analysis

```python
#correlation matrix
print("\nCorrelation Matrix:")
corr_matrix=np.round(df.corr(),2)
corr_matrix
```

```python
print(f"\nCorrelation Heatmap:")
plt.figure(figsize=(8, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', square=True)
plt.title('Correlation Heatmap')
plt.show()
```
